```python
A = SortFile["Total Profit"]
```

```python
## Merge sort algorithm

def MergeSort(A):
    if len(A)>1:
        mid = len(A)//2
        L = A[:mid]
        R = A[mid:]

        MergeSort(L)
        MergeSort(R)

        i=0
        j=0
        k=0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                A[k]=L[i]
                i=i+1
            else:
                A[k]=R[j]
                j=j+1
            k=k+1

        while i < len(L):
            A[k]=L[i]
            i=i+1
            k=k+1

        while j < len(R):
            A[k]=A[j]
            j=j+1
            k=k+1

MergeSort(list(A))
print (A)
```
